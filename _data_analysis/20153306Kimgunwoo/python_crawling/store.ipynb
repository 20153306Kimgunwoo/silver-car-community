```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select

driver = webdriver.Chrome()

url = 'https://www.nhis.or.kr/nhis/policy/retrieveAssistingDevicesRegStoreList.do'
driver.get(url)

pum_mok_cd_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#pumMokCd')))

select = Select(pum_mok_cd_element)

select.select_by_value('17051705')

last_page_number = 201
data_list=[]
for page_number in range(1, last_page_number + 1):
    link_selector = f'a[onclick="fn_retrieveAssistingDevicesRegStoreList({page_number}); return false;"]'
    link = driver.find_element(By.CSS_SELECTOR, link_selector)
    link.click()

    # 페이지 로딩을 기다림 (충분히 장시간이 걸릴 경우 적절히 조절 필요)
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'body')))

    strong_tags = driver.find_elements(By.CSS_SELECTOR, 'tbody strong')
    td_colspans = driver.find_elements(By.CSS_SELECTOR, 'td[colspan="13"]')
    for strong_tag, td_colspan in zip(strong_tags,td_colspans):
        data_dict = {
            '업체':strong_tag.text,
            '주소':td_colspan.text
        }
        data_list.append(data_dict)

driver.quit()
```
